# Chest pain analysis 

A wide range of chest pain is associated to cases of heart failure. These pains will be analyzed according to their problems

```python
rawData.cp.value_counts()
```

rawData.cp means that we are reading the data from the cp column
.value_counts()  It is a panda function used to get (returns) a Series containing counts of unique values. 


We have four types of chest pains and for O we have a number of values and so on.

```python
plt.figure(figsize=(10, 5))
sns.countplot(rawData.cp)
sns.set_style("dark")
plt.xlabel('Chest Type')
plt.ylabel('Count')
plt.title('Chest Type vs Count State')
plt.show()
#0 indicating atypical angina
#1 indicating atypical angina
#2 indicating non-anginal pain
#3 indicating an asymptomatic patient
```

sns.countplot(rawData.cp)- is used  function for plotting the count plot; in this case it is the chest pain column

sns.set_style("dark")- is used to set the theme of the plot in dark style

plt- here we use it to create a bar plot, with the x-axis named Chest Type, y-axis named count and the title of the plot being Chest Type vs Count State

plt.figure(figsize=(10, 5))- creates a plot with the size of 10in by 5in

show()- produces an interactive plot on our screen

target=0-absence of heart disease; 1= heart disease present

```python
cp_zero_target_zero=len(rawData[(rawData.cp==0)&(rawData.target==0)])
cp_zero_target_one=len(rawData[(rawData.cp==0)&(rawData.target==1)])
```

cp_zero_target_zero means when cp=0 & target=0
cp_zero_target_one means when cp=0 & target=1

len- returns the length of an object

```python
sns.barplot(x=['cp_zero_target_zero','cp_zero_target_one'],y=[cp_zero_target_zero,cp_zero_target_one])
sns.set_style("white")
plt.show()
```

sns.barplot- creates a bar plot where each bar represents a summary statistic for each category.
sns.set_style("white")- shows the plot in white theme

The data for the y-axis isn't needed so we don't quote it
